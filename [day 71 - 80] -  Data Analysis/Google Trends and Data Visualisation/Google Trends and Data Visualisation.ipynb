### Bitcoin (BTC) Price v.s. Search Volume

**Challenge**: Create the same chart for the Bitcoin Prices vs. Search volumes. <br>
1. Modify the chart title to read 'Bitcoin News Search vs Resampled Price' <br>
2. Change the y-axis label to 'BTC Price' <br>

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ----- 圖表基本設定 ----- #

plt.figure(
    figsize=(10,6),  # 設定表格大小
    dpi=100  # 設定每英吋點陣數量 (Dots Per Inch)
)

plt.title(
    "Bitcoin News Search vs Resampled Price",  # 設定圖表標題名稱
    fontsize=20  # 設定標題名稱大小
) 


# 設定背景網格

plt.grid(
    color='grey',  # 設定網格顏色
    linestyle='--'  # 設定網格線條圖樣
)


# ----- 圖表邊框設定 (共用參數) ----- #


# 兩軸刻度 => 共用參數設定
plt.yticks(
    fontsize=14  # y軸刻度字體大小
)
plt.xticks(
    fontsize=14,  # x軸刻度字體大小
    rotation=45  # x軸刻度文字旋轉
)


# 共用軸設定
ax1 = plt.gca()  # 取得兩軸資訊 get current axis
ax2 = ax1.twinx()  # 另建立一參數共用x軸



# ----- 圖表邊框設定 (個別參數) ----- #


# y軸邊框名稱
ax1.set_ylabel('Search Trend', color="royalblue", fontsize=14)
ax2.set_ylabel('BTC Price', color="#E6232E", fontsize=14)


# y軸邊框上下限
ax1.set_xlim([df_btc_search.MONTH.min(), df_btc_search.MONTH.max()])
ax2.set_ylim([0,15000])


# ----- 圖表邊框設定 (共用軸刻度設定) ----- #


# 設定大刻度
ax1.xaxis.set_major_locator(
    mdates.YearLocator()
)

# 設定小刻度
ax1.xaxis.set_minor_locator(
    mdates.MonthLocator()
)

# 設定大刻度的標準化顯示
ax1.xaxis.set_major_formatter(
    mdates.DateFormatter('%Y')
)


# ----- 折線圖繪製 ----- #


ax1.plot(
    df_btc_search.MONTH, 
    df_btc_search.BTC_NEWS_SEARCH, 
    color="royalblue", 
    linewidth=3,  # 線條粗細程度
    marker="o"  # 轉折點樣式
)

ax2.plot(
    df_btc_monthly.index, 
    df_btc_monthly.CLOSE, 
    color="#E6232E", 
    linewidth=3,  # 線條粗細程度
    linestyle='dashed'  # 線條樣式
)


plt.show()
```

```python
plt.figure(figsize=(14,10), dpi=100)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', 
          fontsize=20)
plt.grid(color='grey', linestyle='--')


# Increase the size and rotate the labels on the x-axis
plt.yticks(fontsize=14)
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()


# format the ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_minor_locator(months)
ax1.xaxis.set_major_formatter(years_fmt)

ax1.set_ylabel('Search Trend', color="royalblue", fontsize=14)
ax2.set_ylabel('FRED U/E Rate', color="#E6232E", fontsize=14)

ax2.set_ylim([3,10.5])
ax1.set_xlim([df_unemployment.MONTH.min(), df_unemployment.MONTH.max()])


# Calculate the rolling average over a 6 month window
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()


ax1.plot(df_unemployment.MONTH, roll_df.UE_BENEFITS_WEB_SEARCH, 
         color="royalblue", linewidth=3)
ax2.plot(df_unemployment.MONTH, roll_df.UNRATE, 
         color="#E6232E", linewidth=3, linestyle='dashed')


plt.show()
```
